```python
import io
import ffmpeg
from pytube import YouTube
from IPython.display import display, Audio
from google.colab import files
```

```python
# Function to convert YouTube video to audio
def convert_to_audio(video_url):
    try:
        # Load the YouTube video
        youtube_video = YouTube(video_url)
        print(f"Video title: {youtube_video.title}")

        # Filter to get only audio streams and select the first one
        streams = youtube_video.streams.filter(only_audio=True)
        stream = streams.first()

        # Download the audio stream to memory
        audio_data = io.BytesIO()
        stream.stream_to_buffer(audio_data)
        audio_data.seek(0)

        # Save the audio file
        audio_file_path = f"{youtube_video.title}.mp4"
        with open(audio_file_path, 'wb') as audio_file:
            audio_file.write(audio_data.read())

        # Display the audio player
        display(Audio(filename=audio_file_path))

        # Provide download link
        print(f"Download the audio file: [Download]({audio_file_path})")

    except Exception as e:
        print(f"Error: {e}")

# Input YouTube video URL from user input
video_url = input("Enter YouTube video URL:")

# Check if the input is a valid YouTube URL
if video_url.startswith("https://www.youtube.com/") or video_url.startswith("https://youtu.be/"):
    convert_to_audio(video_url)
else:
    if video_url:
        print("Invalid YouTube URL. Please enter a valid URL.")
```

# **Download Youtube video**

```python
# Function to download YouTube video
def download_video(video_url):
    try:
        # Load the YouTube video
        youtube_video = YouTube(video_url)
        print(f"Video title: {youtube_video.title}")

        # Filter to get video streams and select the first one
        streams = youtube_video.streams.filter(file_extension='mp4', res='360p')
        stream = streams.first()

        # Download the video stream to memory
        video_data = io.BytesIO()
        stream.stream_to_buffer(video_data)
        video_data.seek(0)

        # Save the video file with a custom name
        custom_name = "YTvideo.mp4"
        video_file_path = custom_name
        with open(video_file_path, 'wb') as video_file:
            video_file.write(video_data.read())

        # Provide download link
        print(f"Download the video file: [Download]({video_file_path})")

        # Optionally, display a link to download the file
        display(files.download(video_file_path))

    except Exception as e:
        print(f"Error: {e}")

# Get YouTube video URL from user input
video_url = input("Enter YouTube video URL:")

# Check if the input is a valid YouTube URL
if video_url.startswith("https://www.youtube.com/") or video_url.startswith("https://youtu.be/"):
    download_video(video_url)
else:
    if video_url:
        print("Invalid YouTube URL. Please enter a valid URL.")
```

```python
from moviepy.video.io.VideoFileClip import VideoFileClip

# Function to convert MP4 video to audio
def convert_video_to_audio(video_path, audio_path):
    video_clip = VideoFileClip(video_path)
    audio_clip = video_clip.audio
    audio_clip.write_audiofile(audio_path, codec='mp3')
    audio_clip.close()

# Provide the path to your MP4 video and audio file name
video_path = "/content/YTvideo.mp4"  # Replace with your video file path
audio_path = "/content/YTaudio.mp4"  # Replace with your desired audio file path

# Convert video to audio
convert_video_to_audio(video_path, audio_path)

# Display a link to download the audio file
from google.colab import files
files.download(audio_path)
```
