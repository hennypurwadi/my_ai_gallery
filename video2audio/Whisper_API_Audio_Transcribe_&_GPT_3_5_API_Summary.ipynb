```python
import os
import pprint
import openai
openai.api_key = input("Enter your OpenAI API key: ")
```

## Transcribe

```python
audio_file = open("/content/Q-star.mp4", "rb")
```

```python
transcript = openai.Audio.transcribe("whisper-1", audio_file)
```

```python
#print(transcript['text'])
```

```python
transcript_text = transcript['text']
line_length = 150
# Split the text into lines of the specified length
lines = [transcript_text[i:i+line_length] for i in range(0, len(transcript_text), line_length)]

# Print each line separately
for line in lines:
    print(line)
```

## Summarize

```python
response = openai.ChatCompletion.create(
  model="gpt-3.5-turbo-1106",
  messages=[
        {"role": "system", "content": "Q-star"},
        {"role": "user", "content": f"Summarize the following transcript into key bullet points:\n{transcript['text']}"}
    ]
)

print(response['choices'][0]['message']['content'])
```
